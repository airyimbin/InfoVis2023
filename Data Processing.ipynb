## Data Key

For the Anthropographic column, 1 indicates static or non-anthropomorphized, and 2 indicates anthropomorphized.

For the Fear column, 1 indicates the Tourism or Neutral domain, and 2 indicates the Disease or Fearful domain.

For the Medium column, 1 indicates a 2D visualization, and 2 indicates a 3D visualization.

For the Trend column, 1 indicates a positive projected trend, 2 indicates a neutral projected trend, and 3 indicates a negative projected trend.

```python
def parseData(dataset2D, dataset3D):
    allData = pd.DataFrame(columns=['PID', 'Anthropographic', 'Fear', 'Medium', 'Trend','Response'])
    for i in range(dataset2D.shape[0]):
        pid = i
        for j in range(dataset2D.shape[1]):
            columnName = dataset2D.columns[j]
            
            anthro = 2
            if (('Dax' in columnName) or ('Osip' in columnName) or ('Gazzer' in columnName)):
                anthro = 1
                
            fear = 1
            if ('Disease' in columnName):
                fear = 2
                
            trend = 1
            if (('Wug' in columnName) or ('Osip' in columnName)):
                trend = 3
            elif (('Fip' in columnName) or ('Gazzer' in columnName)):
                trend = 2
            
            row = {'PID':pid, 'Anthropographic':anthro, 'Fear':fear, 'Medium': 1, 'Trend': trend, 'Response':dataset2D.iloc[i,j]}
            allData = allData.append(row, ignore_index=True)    
    pid=pid+1
    for i in range(dataset3D.shape[0]):
        pid = i + pid
        for j in range(dataset3D.shape[1]):
            columnName = dataset3D.columns[j]
            
            anthro = 2
            if (('Dax' in columnName) or ('Osip' in columnName) or ('Gazzer' in columnName)):
                anthro = 1
                
            fear = 1
            if ('Disease' in columnName):
                fear = 2
                
            trend = 1
            if (('Wug' in columnName) or ('Osip' in columnName)):
                trend = 3
            elif (('Fip' in columnName) or ('Gazzer' in columnName)):
                trend = 2
                
            row = {'PID':pid, 'Anthropographic': anthro, 'Fear': fear, 'Medium': 2, 'Trend': trend, 'Response':dataset3D.iloc[i,j]}
            allData = allData.append(row, ignore_index=True)   
    return allData
```

```python
def parseDataCat(dataset2D, dataset3D):
    allData = pd.DataFrame(columns=['PID', 'Anthropographic', 'Fear', 'Medium', 'Trend', 'Response'])
    for i in range(dataset2D.shape[0]):
        pid = i
        for j in range(dataset2D.shape[1]):
            columnName = dataset2D.columns[j]
            
            anthro = "Anthropomorphized"
            if (('Dax' in columnName) or ('Osip' in columnName) or ('Gazzer' in columnName)):
                anthro = "Static"
                
            fear = "Neutral"
            if ('Disease' in columnName):
                fear = "Fearful"
            
            trend = "Positive"
            if (('Wug' in columnName) or ('Osip' in columnName)):
                trend = "Negative"
            elif (('Fip' in columnName) or ('Gazzer' in columnName)):
                trend = "Neutral"
                
            row = {'PID':pid, 'Anthropographic':anthro, 'Fear':fear, 'Medium': "2D", 'Trend': trend, 'Response':dataset2D.iloc[i,j]}
            allData = allData.append(row, ignore_index=True)    
    pid=pid+1
    for i in range(dataset3D.shape[0]):
        pid = i + pid
        for j in range(dataset3D.shape[1]):
            columnName = dataset3D.columns[j]
            
            anthro = "Anthropomorphized"
            if (('Dax' in columnName) or ('Osip' in columnName) or ('Gazzer' in columnName)):
                anthro = "Static"
                
            fear = "Neutral"
            if ('Disease' in columnName):
                fear = "Fearful"
            
            trend = "Positive"
            if (('Wug' in columnName) or ('Osip' in columnName)):
                trend = "Negative"
            elif (('Fip' in columnName) or ('Gazzer' in columnName)):
                trend = "Neutral"
                
            row = {'PID':pid, 'Anthropographic':anthro, 'Fear':fear, 'Medium': "3D", 'Trend': trend, 'Response':dataset3D.iloc[i,j]}
            allData = allData.append(row, ignore_index=True)   
    return allData
```
